```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("../datasets/inpatient_data.csv")
```

```python
# Browse all data
dataset
```

```python
# Select one row
row = dataset.loc[0]
row
```

```python
# Select one column
provider_ids = dataset['Provider Id']
provider_ids
```

```python
# Select only certain rows
payments = dataset[['Provider State','Average Total Payments','Average Medicare Payments']]
payments
```

```python
# Reference a single cell
payments.loc[0,'Provider State']
```

```python
# Do math on columns
savings = payments[:]
savings['Average Medicare Savings'] = payments['Average Total Payments'] - payments['Average Medicare Payments']
savings
```

```python
# Filter values
savings[savings['Average Medicare Savings'] > 1000]
```

```python
# Average values by state, then sort by average total payments from highest to lowest
state_averages = savings.groupby('Provider State').mean()
state_averages = state_averages.sort_values('Average Total Payments', ascending=False)
state_averages
```

```python
# Let's do a similar thing, but also calculate std, then sort by average total payments from highest to lowest
state_stats = savings.groupby('Provider State').agg(
    {'Average Total Payments': ['mean','std'],
     'Average Medicare Payments': ['mean','std'],
     'Average Medicare Savings': ['max','min']})
state_stats
```
